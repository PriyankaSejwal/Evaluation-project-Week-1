Feature XLarge Bags have almost half of the value as 0. We will deal with these using the median as XLarge Bags column has outliers and is skewed too so mean is not a great option to replace the zeroes with.

```python
from sklearn.impute import SimpleImputer 
```

```python
si=SimpleImputer(missing_values=0.00,strategy='median')
```

```python
avocado['XLarge Bags']=si.fit_transform(avocado['XLarge Bags'].values.reshape(-1,1))
```

```python
sum(avocado['XLarge Bags']==0)
```

```python
avocado
```

```python
sum(avocado['Large Bags']==0)
```

we can see that there are null nalues in 'Large Bags' too. We will replace it with median as there is skewness and outliers in this feature too.

```python
avocado['Large Bags']=si.fit_transform(avocado['Large Bags'].values.reshape(-1,1))
```

```python
sum(avocado['Large Bags']==0)
```

We have successfully removed the zero values from the data.

```python
avocado.skew()
```

```python
#Removing skewness using log transform
for i in avocado.columns:
    if avocado.skew().loc[i]>0.55:
        avocado[i]=np.log1p(avocado[i])
```

```python
avocado.skew()
```

```python
#Removing outliers
from scipy.stats import zscore
```

```python
z=np.abs(zscore(avocado))
```

```python
threshold=3
avocado_new=avocado[(z<3).all(axis=1)]
print(avocado.shape)
print('\n')
print('After removing outliers shape comes down to:')
print(avocado_new.shape)
```

Separating the target variable- Average Price

```python
x=avocado_new.drop(['AveragePrice'],axis=1)
```

```python
x
```

```python
y=avocado_new['AveragePrice']
```

```python
y
```

```python
#applying standard scaler to dependant variable.
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_new=sc.fit_transform(x)
x=pd.DataFrame(x_new,columns=x.columns)
```

```python
x
```

```python
x.skew()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=42)
```

```python
print('Shape of x train: ',x_train.shape)
print('Shape of y Train: ',y_train.shape)
print('Shape of x test: ',x_test.shape)
print('Shape of y test: ',y_test.shape)
```

```python
list=[LinearRegression(),Lasso(),Ridge(),ElasticNet(),DecisionTreeRegressor(),SVR(),KNeighborsRegressor()]
for i in list:
    i.fit(x_train,y_train)
    predict=i.predict(x_test)
    print('Score of: ',i,'is: ',i.score(x_train,y_train))
    print('R2 Score corresponding to: ',i,'is: ',r2_score(y_test,predict))
    print('Mean Absolute Error: ',mean_absolute_error(y_test,predict))
    print('Mean Squared Error: ',mean_squared_error(y_test,predict))
    print('Root Mean Sqaured error: ',np.sqrt(mean_squared_error(y_test,predict)))
    print('\n')
    print('*********************************************************')
    print('\n')
```

SVR is the best  model as r2 score for SVR is higher than others with a less Mean absolute error, mean squared error, root mean squared error. finding its best parameters with the help of grid search cv

```python
from sklearn.model_selection import GridSearchCV
svr=SVR()
parameters={ 'kernel':['linear','poly','rbf']}
cv=GridSearchCV(svr,parameters)
cv.fit(x,y)
cv.best_params_
```

```python
# Using SVR with the best parameters.
svr=SVR(kernel='rbf')
svr.fit(x_train,y_train)
predict=svr.predict(x_test)
print('Score is: ',svr.score(x_train,y_train))
print('R2 Score corresponding is: ',r2_score(y_test,predict))
print('Mean Absolute Error: ',mean_absolute_error(y_test,predict))
print('Mean Squared Error: ',mean_squared_error(y_test,predict))
print('Root Mean Sqaured error: ',np.sqrt(mean_squared_error(y_test,predict)))
print('\n')
print('*********************************************************')
print('\n')
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
score=cross_val_score(svr,x,y,cv=4,scoring='r2')
print('Score: ',score)
print('Mean Score: ',score.mean())
print('Standard Deviation: ',score.std())
```

```python
#Using adaboost to boost our model svr.
from sklearn.ensemble import AdaBoostRegressor
ada=AdaBoostRegressor(base_estimator=svr,n_estimators=15,learning_rate=0.1,random_state=42)
ada.fit(x_train,y_train)
pred=ada.predict(x_test)
print('Score: ',ada.score(x_train,y_train))
print('Mean absolute error: ',mean_absolute_error(y_test,pred))
print('Mean squared error: ',mean_squared_error(y_test,pred))
print('Root mean squared error:',np.sqrt(mean_squared_error(y_test,pred)))
print('R2 score: ',r2_score(y_test,pred))
```

```python
#Using ensemble technique to boost our score
import numpy as np
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=20,random_state=42)
rf.fit(x_train,y_train)
print('Score of random forest is:',rf.score(x_train,y_train))
predict=rf.predict(x_test)
print('Mean absolute error: ',mean_absolute_error(y_test,predict))
print('Mean squared error: ',mean_squared_error(y_test,predict))
print('Root mean squared error:',np.sqrt(mean_squared_error(y_test,predict)))
print('R2 score: ',r2_score(y_test,predict))
```

```python
from sklearn.model_selection import GridSearchCV
rf=RandomForestRegressor()
parameters={'criterion':['mse','mae'],'random_state':range(42,60)}
cv=GridSearchCV(rf,parameters)
cv.fit(x,y)
cv.best_params_
```

```python
#Cross Validation for random forest regressor
rf=RandomForestRegressor(criterion='mae',random_state=54)
score=cross_val_score(rf,x,y,cv=3,scoring='r2')
print('Score: ',score)
print('Mean Score: ',score.mean())
print('Standard Deviation: ',score.std())
```

Working on the best model i.e. Random forest regressor as random forest regressor has the highest r2 score and the errors are quite low.

```python
rfr=RandomForestRegressor(criterion='mae',n_estimators=20,random_state=54)
rfr.fit(x_train,y_train)
print('Score of random forest is:',rfr.score(x_train,y_train))
predict=rfr.predict(x_test)
print('Mean absolute error: ',mean_absolute_error(y_test,predict))
print('Mean squared error: ',mean_squared_error(y_test,predict))
print('Root mean squared error:',np.sqrt(mean_squared_error(y_test,predict)))
print('R2 score: ',r2_score(y_test,predict))
```

```python
a=[DecisionTreeClassifier(),SVC(),KNeighborsClassifier()]
for j in a:
    max_r_score=0
    print('for the algorithm :',j)
    for i in range(33,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=.25)
        j.fit(x_train,y_train)
        pred=j.predict(x_test)
        score=j.score(x_train,y_train)
        accuracy=accuracy_score(pred,y_test)
        print('accuracy score corresponding to random state',i,'is: ',accuracy)
        if accuracy>max_r_score:
            max_r_score=accuracy
            final=i
    print('\n')
    print('Max accuracy score corresponding to: ', final,'is: ',max_r_score)
    print('\n')
        
```
